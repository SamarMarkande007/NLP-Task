#Assignment 1

Removing Punctuation, removing Stopwords, Stemming and Lemmatization.

```python
import nltk
nltk.download('punkt_tab')
```

```python
text = """Football is an intense, fast-paced sport! Players run, pass, tackle, and score goals; each action requires precision, skill, and teamwork. A striker shoots, a defender blocks, and a goalkeeper saves — all within seconds. The crowd cheers loudly: “Goal!” when the ball hits the net. Isn’t it exciting?

In modern football, strategy is crucial: coaches analyze opponents, plan formations, and adjust tactics. Teams may attack aggressively, defend deeply, or control possession carefully. However, mistakes — such as poor passes, missed chances, or defensive errors — can change the outcome quickly. Therefore, discipline, focus, and coordination are essential for success."""
```

#  Removing Punctuation

```python
import re
text_clean = re.sub(r'[^\w\s]', '', text)
print(text_clean)
```

# Tokenization

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text_clean.lower())
print(tokens)
```

# Removing Stopwords

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

filtered_words = [word for word in tokens if word.lower() not in stop_words]
print(filtered_words)
```

# Stemming

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()

stemmed_words=[stemmer.stem(word) for word in filtered_words]
print(stemmed_words)
```

# Lemmatization

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()

lemmatized_words=[lemmatizer.lemmatize(word,pos='v') for word in filtered_words]
print(lemmatized_words)
```

```python
converted_text=" ".join(lemmatized_words)
print(converted_text)
```

# Assignment 2

Bag of Words, TF-IDF, Word2Vec and Checking Similarity.

# Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = [converted_text]

vectorizer = CountVectorizer()
vectorized_text = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())
print(vectorized_text.toarray())
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
corpus_1 = [converted_text]

vectorizer_1 = TfidfVectorizer()
vectorized_text_1 = vectorizer_1.fit_transform(corpus_1)

print(vectorizer_1.get_feature_names_out())
print(vectorized_text_1.toarray())
```

# Large text and Preprocessing

```python
large_text = """Football, also known as soccer, is a popular sport played worldwide. Players compete, battle, and struggle to control the ball and score goals. A team must cooperate, collaborate, and coordinate effectively to win matches. The striker attempts to shoot, score, and finish opportunities, while midfielders pass, distribute, and create chances. Defenders protect, guard, and block attacks, and goalkeepers save, stop, and prevent goals.

Winning a match requires effort, hard work, and dedication. Teams train, practice, and prepare regularly to improve performance. A good player must be quick, fast, and rapid in movement, while also being strong, powerful, and resilient. Coaches guide, instruct, and mentor players to follow strategies and tactics.

Sometimes teams dominate, control, and manage possession, while others defend, resist, and counterattack. Matches can be exciting, thrilling, and exhilarating, especially when goals are scored in the final minutes. Fans support, cheer, and encourage their favorite teams passionately.

Football is not just a game; it is an emotion, a passion, and a feeling shared by millions. Victory brings joy, happiness, and celebration, while defeat causes sadness, disappointment, and frustration. Despite challenges, teams continue to fight, compete, and strive for success."""
```

```python
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(large_text.lower())
print(sentences)
```

```python
import re
clean_sentences = [re.sub(r'[^\w\s]', '', sentence) for sentence in sentences]
print(clean_sentences)
```

```python
sentences_tokens = [word_tokenize(sentence) for sentence in clean_sentences]
print(sentences_tokens)
```

# Word2Vec

```python
!pip install gensim
```

```python
from gensim.models import Word2Vec
model = Word2Vec(
    sentences_tokens,
    vector_size = 100,
    window = 5,
    min_count = 1,
    epochs = 100
)
```

# Checking Similarity

```python
print("Similarity between fast and quick: ",model.wv.similarity('fast', 'quick'))
```

```python
print("Similarity between joy and happiness: ",model.wv.similarity('joy','happiness'))
```

```python
print("Similarity between prepare and practice: ",model.wv.similarity('prepare','practice'))
```

```python
print("Similarity between sadness and disappointment: ",model.wv.similarity('sadness','disappointment'))
```
